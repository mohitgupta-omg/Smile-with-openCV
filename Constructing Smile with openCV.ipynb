```python
# import libraries
import cv2
import numpy as np
```

```python
''' Making Smiley '''
img = np.zeros((620, 620, 3), np.uint8)
```

```python
# text
font = cv2.FONT_ITALIC
cv2.putText(img,'bE hApPy :)',(56,579), font, 3, (51,255,51), 4)
```

```python
# Hat
cv2.line(img, (270,82), (460, 152), (102, 102, 255), 14)
cv2.line(img, (320,72), (440, 117), (102, 102, 255), 14)
cv2.line(img, (360,56), (415, 76), (102, 102, 255), 14)
```

```python
# Face
cv2.circle(img, (300, 270), 160, (250, 250, 0), 5)
```

```python

# Eyes
cv2.circle(img, (230, 215), 40, (250, 250, 0), 3)
cv2.circle(img, (362, 215), 40, (250, 250, 0), 3)

cv2.circle(img, (230, 215), 20, (102, 102, 255), -1)
cv2.circle(img, (362, 215), 20, (102, 102, 255), -1)
```

```python
# Nose
cv2.line(img, (300,250), (300, 310), (255, 255, 0), 5)
```

```python
# Lips
cv2.ellipse(img,(300,310), (90,40), 0, 0 , 180, (255, 255, 0), 4)
cv2.ellipse(img,(300,332), (70,60), 0, 0 , 180, (255, 255, 0), 4)
```

```python
cv2.imshow('Keep Smiling with Python.py', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
